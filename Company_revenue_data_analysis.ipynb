# **Data analysis**

Revenue by Industry

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 'Revenue' is the column name in the dataframe

data['Revenue'] = data['Revenue'].replace('[\$,]', '', regex=True).astype(float)

# Use seaborn to create a bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='Industry', y='Revenue', data=data, palette='tab10', ci=None)  # 'tab10' is a colormap with 10 distinct colors

plt.xlabel('Industry', fontsize=15)
plt.ylabel('Revenue in USD', fontsize=15)
plt.title('Revenue per Industry', fontsize=20)
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility

plt.show()
```

Profit by industry

Revenue vs Profit

```python
import numpy as np
import matplotlib.pyplot as plt


revenue = np.array(data.Revenue)
profit = np.array(data.Profit)
sort_rev = np.argsort(revenue)
sort_profit = np.argsort(profit)
index_1 = np.arange(len(revenue))  # Create an index array
index_2 = np.arange(len(profit))

# Scatter plot for Revenue
plt.figure(figsize=(10, 10))  # Create a new figure for the Revenue
plt.scatter(index_1, sort_rev, color='blue')  # Scatter plot index vs. revenue
plt.title("Revenue Over Time")  # Title of the graph
plt.xlabel('Quantity', fontsize=13)  # Label for the x-axis
plt.ylabel('Revenue', fontsize=13)  # Label for the y-axis
plt.show()  # Show the plot

# Scatter plot for Profit
plt.figure(figsize=(10, 10))  # Create a new figure for the Profit
plt.scatter(index_2, sort_profit, color='green')  # Scatter plot index vs. profit
plt.title("Profit Over Time")  # Title of the graph
plt.xlabel('Quantity', fontsize=13)  # Label for the x-axis
plt.ylabel('Profit', fontsize=13)  # Label for the y-axis
plt.show()  # Show the plot
```

**Revenue vs Quantity:** The graph shows a linear relationship between the two variables, i.e as quantity increases over time, revenue decreases at the same rate, resulting in a indirectly proportional relationship. Suggesting that factors such as volume discounts or lower-priced items selling in higher volumes.

**Profit vs Quantity:** The graph doesn't show a clear trend or correlation between profit and quantity, since the data points appear to be dispersed. This suggests that profit doesn't consistently increase or decrease with changes in quantity which can depend on the type of goods and services which are being provided to consumers. Profit can also be influenced by factors such as cost of goods sold, operational efficies, or different profit margins across different products.

# **Assumptions which can be made:**

*   The data could possibly represent different goods or services with varying profit margins and pricing strategies, explaining the clear pattern in revenue versus the scattered pattern in profit.
*   The first graph could also suggest that a fixed pricing strategy that adjusts based on the volume of goods sold, or possibly strategies that reduce unit price at higher quantites like selling goods/services at a wholesale price like Costco.









Market share by revenue

```python
import random
company_revenue = data.groupby('Name')['Revenue'].sum().head(10)


# Creating a pie chart for the revenue distribution among companies
plt.figure(figsize=(16, 20))
plt.pie(company_revenue, labels=company_revenue.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue Distribution by Company')

# Find the maximum revenue value
max_revenue = company_revenue.max()

# Display the maximum value
print("Maximum Revenue:", max_revenue)

# Find the company with the maximum revenue
company_with_max_revenue = company_revenue.idxmax()

# Display the company with the maximum revenue
print("Company with Maximum Revenue:", company_with_max_revenue)

plt.show()
```

Based on the pie chart above, it shows the top 10 companies with the highest revenue which is Amazon of $574785.0.

```python
import random
company_revenue = data.groupby('Name')['Revenue'].sum().tail(10)


# Creating a pie chart for the revenue distribution among companies
plt.figure(figsize=(16, 20))
plt.pie(company_revenue, labels=company_revenue.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue Distribution by Company')

# Find the minimum revenue value
min_revenue = company_revenue.min()

# Display the minimum value
print("Minimum Revenue:", min_revenue)

# Find the company with the minimum revenue
company_with_min_revenue = company_revenue.idxmin()

# Display the company with the minimum revenue
print("Company with Minimum Revenue:", company_with_min_revenue)

plt.show()
```

Based on the other pie chart above, it is clear that from the bottom 10 companies with the lowest revenue is Valero Energy of $171189.0

```python
# Convert currency formatted strings to numbers for 'Revenue' and 'Revenue per worker'
data['Revenue'] = data['Revenue'].replace('[\$,]', '', regex=True).astype(float)
data['Revenue per worker'] = data['Revenue per worker'].replace('[\$,]', '', regex=True).astype(float)

# Ensure 'Employees' is also an integer for correlation calculation
data['Employees'] = data['Employees'].astype(int)

# Industry Comparison: Calculate the average revenue per worker by industry
average_rev_per_worker_by_industry = data.groupby('Industry')['Revenue per worker'].mean()

# Calculate correlation between the number of employees and revenue per worker
correlation = data['Employees'].corr(data['Revenue per worker'])

# Rank companies based on revenue per worker in descending order
ranked_companies = data.sort_values('Revenue per worker', ascending=True)

# Display results
print("Average Revenue per Worker by Industry:\n", average_rev_per_worker_by_industry)
print("\nCorrelation between Employees and Revenue per Worker:", correlation)
print("\nCompanies Ranked by Revenue per Worker:\n", ranked_companies[['Name', 'Revenue per worker']])
```

The correlation between employees and revenue per worker is negative which suggests by having a higher number of employees it doesn't always gurantee a higher revenue per worker as this can depend on the industry the company is in.

Vitol has a revenue per worker of $323,717,948.72 despite having only 1560 employees. Which shows that companies with less number of employees are able to generate more revenue (per worker) as they are increasing their efficiency.
